```python
#read data and transfer the data into matrix
def read_data(file_path):
    max_user_id = 0
    max_movie_id = 0
    
    with open(file_path, 'r') as file:
        for line in file:
            #split each row data in train text and map the values to matrix
            user_id, movie_id, rating = map(int, line.split())
            if user_id > max_user_id:
                max_user_id = user_id
            if movie_id > max_movie_id:
                max_movie_id = movie_id
    
    # create a matrix of U-M and set dimension to (max_user_id, max_movie_id)
    user_movie_matrix = np.zeros((max_user_id, max_movie_id))
    
    # fill the matrix
    with open(file_path, 'r') as file:
        for line in file:
            user_id, movie_id, rating = map(int, line.split())
            user_movie_matrix[user_id-1, movie_id-1] = rating
    
    return user_movie_matrix, max_user_id, max_movie_id

train_file_path = "E:\\文档\\AA-SCU\\272\\Proj 2\\train.txt"
user_movie_matrix, max_user_id, max_movie_id = read_data(train_file_path)
# transfer the numpy list into dataframe
df = pd.DataFrame(user_movie_matrix, columns=[str(i) for i in range(user_movie_matrix.shape[1])])

```

```python
#Cosine Similarity
def updated_cos_sim(v1, v2):
    v1_new = []
    v2_new = []
    for index in range(len(v1)):
        if v1[index] != 0 and v2[index] != 0:
            v1_new.append(v1[index])
            v2_new.append(v2[index])
    #No similar dimenions between users, set sim==0
    if(len(v1_new) == 0 and len(v2_new) == 0):
        sim = 0
        return sim
    """
    #Only 1 dimension that is similar between the two users
    #set sim==inverse euclidean distance
    if(len(v1_new) == 1 and len(v2_new) == 1):
        euclidean_distance = np.linalg.norm(v1_new[0]-v2_new[0])
        sim = 1/(1+euclidean_distance)
        return sim

    """

    
    numerator = np.dot(v1_new,v2_new)
    v1_norm = np.sqrt(np.sum(np.square(v1_new)))
    v2_norm = np.sqrt(np.sum(np.square(v2_new)))
    #calculate the cosine similarity 
    sim = numerator/(v1_norm * v2_norm)
    
    return sim
```

```python
#Calculating the cos_sim for each active user with every single user in the training data. 
#Row 0 is the list of similarities between active user 201 with users 1-200
#Matrix Dimensions: 100x200

def create_cos_sim_matrix(matrix, users_arr, training_arr):
    count = 0
    matrix = []
    vals = []
    for key in users_arr:
        for row in training_arr:
            cos_sim_value = updated_cos_sim(users_arr[key],row)
            vals.append(cos_sim_value)
            count += 1
            if count == 200:
                matrix.append(vals)
                vals = []
                count = 0
    return matrix

def print_matrix_dimensions(matrix):
    print("Rows: " + str(len(matrix)))
    print("Columns: " + str(len(matrix[0])) + "\n")
```

```python
#Returns the list of matrix(100x200) of indexes with the highest cosine similarity in ascending order
import heapq
def top_k_sim(matrix,k=200):
    max_index = []
    for users in matrix:# 遍历每个活跃用户的余弦相似度列表
        abs_cos_sim = np.abs(users)  # 计算当前用户所有相似度的绝对值
        top_k_indices = heapq.nlargest(k, range(len(abs_cos_sim)), key=abs_cos_sim.__getitem__)  # 获取相似度最高的 k 个索引
        max_index.append(top_k_indices)  # 将这些索引添加到结果列表中
    return max_index  # 返回包含所有活跃用户相似度最高的 k 个用户索引的列表

top_k_5 = top_k_sim(cos_sim_matrix_5,k=200)
top_k_10 = top_k_sim(cos_sim_matrix_10,k=200)
top_k_20 = top_k_sim(cos_sim_matrix_20,k=200)
```

```python
def weighted_rating_cos_sim(test_file, shift_index, top_k, cos_sim_matrix, default_k, sim_threshold=0.6):
    output = []
    
    for index, row in test_file.iterrows():
        #Only rating when the current rating is 0
        if row[2] == 0: 
            numerator = 0
            denominator = 0
            temp_counter = 0
            user_id = row[0] - shift_index
            movie_id = row[1] - 1
            
            indexes = top_k[user_id]
            reverse_indexes = indexes.copy()
            reverse_indexes.reverse()
            
            for ind in reverse_indexes:
                cos_sim = cos_sim_matrix[user_id][ind]
                ranking = arr[ind][movie_id]  # 假设 arr 是一个全局变量或传入的参数，表示用户-电影评分矩阵
                
                if ranking != 0 and cos_sim > sim_threshold:
                    temp_counter += 1
                    numerator += (cos_sim * ranking)
                    denominator += cos_sim
                    
                    if temp_counter == default_k:
                        break

            if numerator != 0 and denominator != 0:
                final_rank = numerator / denominator
                final_rank = int(round(final_rank, 0))
                line = [row[0], row[1], final_rank]
                output.append(line)
            else:
                line = [row[0], row[1], 3]  # 默认评分为3
                output.append(line)
            
            # 打印用户实际找到的相似用户数量用于调试
            if temp_counter >= 95:
                print(f"User {row[0]} has {temp_counter} similar users considered.")
    
    return output

```

```python
def convert_list_to_str(line):
    string = str(line)[1:-1]
    string = string.replace(',', '')
    return string
```

```python
def calculate_pearson_adjusted_ratings(user_ratings_matrix):
    adjusted_train_ratings_matrix = []  # 存储经过均值调整后的评分矩阵
    train_user_means = []  # 存储每个用户的评分均值

    for user_ratings in user_ratings_matrix:  # 遍历每个用户的评分
        non_zero_ratings = [rating for rating in user_ratings if rating != 0]
        if non_zero_ratings:
            mean_rating = sum(non_zero_ratings) / len(non_zero_ratings)
        else:
            mean_rating = 0
        
        train_user_means.append(mean_rating)
        
        adjusted_ratings = [(rating - mean_rating) if rating != 0 else 0 for rating in user_ratings]
        adjusted_train_ratings_matrix.append(adjusted_ratings)

    return adjusted_train_ratings_matrix, train_user_means
```

```python
def calculate_pearson_adjusted_ratings_for_test(test_user_ratings_dict):
    adjusted_test_ratings_matrix = []
    test_user_means = []

    for user_id, user_ratings in test_user_ratings_dict.items():
        non_zero_ratings = [rating for rating in user_ratings if rating != 0]
        if non_zero_ratings:
            mean_rating = sum(non_zero_ratings) / len(non_zero_ratings)
        else:
            mean_rating = 0
        
        test_user_means.append(mean_rating)
        
        adjusted_ratings = [(rating - mean_rating) if rating != 0 else 0 for rating in user_ratings]
        adjusted_test_ratings_matrix.append(adjusted_ratings)
        
        # 更新用户评分字典
        test_user_ratings_dict[user_id] = adjusted_ratings

    return adjusted_test_ratings_matrix, test_user_means
```

```python
def weighted_rating_pearson_correlation(test_data, shift_index,top_k_sim, cos_sim_matrix, test_user_means, train_user_means, ratings_matrix, k):
    output = []

    for index, row in test_data.iterrows():
        #filtering those unrated movies
        if row[2] == 0:
            active_user_id = row[0] - shift_index
            movie_id = row[1] - 1

            indexes = top_k_sim[active_user_id]
            reverse_indexes = indexes[::-1]

            numerator = 0
            denominator = 0
            counter = 0

            test_mean_val = test_user_means[active_user_id]

            for ind in reverse_indexes:
                mean_val_train_user = train_user_means[ind]
                cos_sim = cos_sim_matrix[active_user_id][ind]  # 修正为直接索引
                rating = ratings_matrix[ind][movie_id]

                if rating != 0:
                    numerator += cos_sim * (rating - mean_val_train_user)
                    denominator += abs(cos_sim)
                    counter += 1

                    if counter == k:
                        break
            

            if denominator != 0:
                final_rank = test_mean_val + (numerator / denominator)
                final_rank = float(np.asscalar(np.array(final_rank)) if isinstance(final_rank, np.ndarray) else final_rank)  # 确保 final_rank 是一个标量
                final_rank = final_rank.item() if isinstance(final_rank, np.ndarray) else final_rank
                final_rank = int(round(final_rank, 0))

                final_rank = min(max(final_rank, 1), 5)
                output.append([row[0], row[1], final_rank])
            else:
                output.append([row[0], row[1], 3])
                

    return output
```

```python
#Calculating the number of movie that has non_zero ratings
def cal_count_training_non_zero():
    movie_ratings_count = [0] * 1000
    # 遍历每个用户的电影评分
    for user_ratings in arr:
        # 遍历每个电影的评分
        for movie_index, rating in enumerate(user_ratings):
            # 如果评分不为0，增加对应电影的评分次数
            if rating != 0:
                movie_ratings_count[movie_index] += 1
    # 打印电影评分次数列表的长度
    print("Length of No_Zero_Ratings_Movie Count: " + str(len(movie_ratings_count)))
    
    return movie_non_zero_ratings_count
```

```python
#Returns an array of iuf values for each training user
def cal_iuf_train_arr(movie_non_zero_ratings_count):
    iuf_train_arr = []
    for i in range(len(movie_non_zero_ratings_count)):
        if movie_non_zero_ratings_count[i] != 0:
            iuf_rating = math.log10((200/movie_non_zero_ratings_count[i]))
        else:
            iuf_rating = 1
        iuf_train_arr.append(iuf_rating)
    print("Length of iuf_training_array: " + str(len(iuf_train_arr)))
    return iuf_train_arr
```

```python
#Returns the training matrix with each rating = rating * IUF value
def applied_iuf_train_matrix(training_data, iuf_training_vals):
    
    adjusted_training_data = training_data.copy()
    for user_index, user_ratings in enumerate(adjusted_training_data):
        for movie_index, rating in enumerate(user_ratings):
            if rating != 0:
                adjusted_training_data[user_index][movie_index] *= iuf_training_vals[movie_index]
    
    return adjusted_training_data
```

```python
#Return the final IUF_training matrix(200x1000)

def final_iuf_train_arr():
    movie_non_zero_ratings_count = cal_count_training_non_zero()
    iuf_train_arr = cal_iuf_train_arr(movie_non_zero_ratings_count)
    #print(iuf_train_arr)

    training_data = arr.copy()
    training_data = training_data.astype(float)
    
    adjusted_training_data = applied_iuf_train_matrix(training_data, iuf_train_arr)
    print("Rows of final_iuf_train_data: " + str(len(adjusted_training_data)))
    print("Cols of final_iuf_train_data: " + str(len(adjusted_training_data[0])))
    return iuf_training_data

iuf_training_data = final_iuf_train_arr()
```

```python
def adjusted_cosine_similarity(ratings1, ratings2, active_user_mean):
    # 初始化相关变量
    filtered_ratings1 = []
    filtered_ratings2 = []
    numerator = 0
    sum_sq_diff1 = 0
    sum_sq_diff2 = 0
    
    # 过滤并计算所需的组件
    for rating1, rating2 in zip(ratings1, ratings2):
        if rating1 != 0 and rating2 != 0:
            diff1 = rating1 - active_user_mean
            diff2 = rating2 - active_user_mean
            filtered_ratings1.append(diff1)
            filtered_ratings2.append(diff2)
            numerator += diff1 * diff2
            sum_sq_diff1 += diff1 ** 2
            sum_sq_diff2 += diff2 ** 2
    
    # 处理特殊情况
    if not filtered_ratings1:  # 如果没有共同评分的用户
        return 0

    if len(filtered_ratings1) == 1:  # 如果只有一个共同评分的用户
        euclidean_distance = np.linalg.norm(filtered_ratings1[0] - filtered_ratings2[0])
        return 1 / (1 + euclidean_distance)
    
    # 计算余弦相似度
    denominator = np.sqrt(sum_sq_diff1 * sum_sq_diff2)
    if denominator == 0:
        return 0
    return numerator / denominator
```

```python
#Creating a matrix with alredy rated movie lists for each active user
#test5(100x5)
#test10(100x10)
#test20(100x20)


def create_test_user_ratings_matrix(num_ratings, test_file):
    total_test_users_rated_movie_list = []
    user_rated_movies = []
    counter = 0
    
    for index,row in test_file.iterrows():
        if row[2] == 0:
            continue;

        movie_id = row[1]-1
        user_rated_movies.append((movie_id,row[2]))
        counter += 1
        if counter == num_ratings:
            total_test_users_rated_movie_list.append(user_rated_movies)
            counter = 0
            user_rated_movies = []
    return total_test_users_rated_movie_list

```

```python
def item_based_weighted_ranking(test_ratings, rated_movies, item_means, ratings_data, offset, top_k):
    results = []

    for _, row in test_ratings.iterrows():
        if row[2] == 0:
            user_id = row[0] - offset
            target_movie_id = row[1] - 1

            cos_sim_weights = []
            for rated_movie in rated_movies[user_id]:
                weight = adjusted_cosine_similarity(ratings_data[rated_movie[0]], ratings_data[target_movie_id], item_means[user_id])
                cos_sim_weights.append(weight)

            if cos_sim_weights:
                # Get the indices of the top k similar items sorted by their similarity weight using the defined function
                top_indices = item_based_top_k(cos_sim_weights, top_k)
                
                numerator = 0
                denominator = 0
                for idx in top_indices:
                    weight = cos_sim_weights[idx]
                    if weight > 0.3:
                        rating = rated_movies[user_id][idx][1]
                        numerator += weight * rating
                        denominator += weight

                if denominator > 0:
                    predicted_rating = round(numerator / denominator)
                    predicted_rating = max(min(predicted_rating, 5), 1)
                else:
                    predicted_rating = 3  # Default rating if no suitable weights are found

                results.append([row[0], row[1], predicted_rating])

    return results

```

```python
from scipy.sparse.linalg import svds
def predict_ratings_svd(test_data, output_filename, shift_index):
    train_matrix=df.copy()
    U, sigma, Vt = svds(train_matrix, k=50)
    sigma = np.diag(sigma)
    all_predicted_ratings = np.dot(np.dot(U, sigma), Vt)

    with open(output_filename, 'w') as file:
        for index, row in test_data.iterrows():
            user_id = row[0]
            movie_id = row[1]
            rating = row[2]

            # 只对原本评分为0的电影进行预测
            if rating == 0:
                user_index = user_id - shift_index
                movie_index = movie_id - 1
                predicted_rating = round(all_predicted_ratings[user_index, movie_index])
                # 保证评分在1到5之间
                if predicted_rating < 1:
                    predicted_rating = 3  # 将不合理的低评分设置为中性评分3
                elif predicted_rating > 5:
                    predicted_rating = 5  # 将高于5的评分限制在5
                file.write(f"{user_id} {movie_id} {predicted_rating}\n")

```
